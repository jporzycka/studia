Uzupełnij poniższe impementacje tak aby osiągnąć zbliżone czasy wykonania
is_perfect: 76.5 ms ± 5.79 ms per loop (mean ± std. dev. of 7 runs, 40 loops each)
is_perfect2: 73 ms ± 4 ms per loop (mean ± std. dev. of 7 runs, 40 loops each)

```python
# Funkcja powinna sprawdzać, czy dana liczba n jest doskonała
# https://pl.wikipedia.org/wiki/Liczba_doskona%C5%82a

def sum_of_dividers(n):
    sum_of_dividers = 0
    for k in range(1,int(n/2)+1):
        if n%k==0:
            sum_of_dividers+=k
    return sum_of_dividers

def is_perfect(n):
    if sum_of_dividers(n) == n:
        return True
    else:
        return False

print(is_perfect(6) is True)
print(is_perfect(28) is True)
print(is_perfect(7) is False)

def sum_of_dividers2(n):
    sum_of_dividers = 0
    for k in range(1,n+1):
        if n%k==0:
            sum_of_dividers+=k
    return sum_of_dividers

def is_perfect2(n):
    if sum_of_dividers2(n) == 2*n:
        return True
    else:
        return False

print(is_perfect2(6) is True)
print(is_perfect2(28) is True)
print(is_perfect2(7) is False)
```

Dla zadanej tablicy o długości n zawierającej liczby całkowite większe od zera. Chcemy znaleźć podtablice, że m kolejnych elementów sumuje się do d.
Napisz funkcję która dla danych m,d  i sekwencję liczb całkowitych większych od zera tab, wypisze na ile sposobów można stworzyć taką podtablice?
Przykładowo, niech m=2, d=3, a czekolada składa się z tab=[1,2,1,3,2]. Chcemy znaleźć takie dwa (bo m=2) elementy pod rząd, których suma da nam d=3. 
Jest to pierwszy i drugi element lub drugi i trzeci.

Napisz program, który działa w czasie liniowym.

```python
def solve(d, m, tab):
    sum = 0
    amount = 0
    first_tab = tab[:m]
    for num in first_tab:
        sum += num
    if sum == d:
        amount += 1
    for i in range(m, len(tab)):
        sum -= tab[i-m]
        sum += tab[i]
        if sum == d:
            amount += 1
    return amount
    
tab=[1, 2, 1, 3, 2]
d=3 
m=2
tab1=[1, 1, 1, 1, 1, 1]
d1=3 
m1=2
tab2=[4,1,3,4,1,1,1,1,1,1,1]
d2=4 
m2=4
print(solve(d, m, tab)==2)
print(solve(d1, m1, tab1)==0)
print(solve(d2, m2, tab2)==4)
```

Napisz definicję funkcji HowManyIntegers(N), która zwróci jak wiele liczb z zakresu [1,N) składa się wyłącznie z cyfr {0,2,7,9}.

Czyli np.

Twoje rozwiązanie powinno mieć jak najmniej niebiałych znaków. Czy uda się w 69? A może da się jeszcze mniej?

```python
import re
def HowManyIntegers(n):
    return sum(1 for x in range(1,n) if re.match('^[0279]+$',str(x)))

print(HowManyIntegers(1) == 0) 
print(HowManyIntegers(10) == 3) # 2,7,9 
print(HowManyIntegers(28) == 6) # 2,7,9,20,22,27
```
